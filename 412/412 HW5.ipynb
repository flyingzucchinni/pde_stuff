### <h2 style="font-family: Times;font-size: 20px">*2.5.4*
Consider $u(x,y)$ satisfying Laplace's equation inside a rectangle $(0<x<L,0<y<H)$ subject to the boundary conditions 
$$\frac{\partial u}{\partial x}(0,y)=0, \space \frac{\partial u}{\partial y}(x,0)=0$$
 $$\frac{\partial u}{\partial x}(L,y)=0, \space \frac{\partial u}{\partial y}(x,H)=f(x)$$
 ### <h2 style="font-family: Times;font-size: 20px">*Solution.*
##### <h2 style="font-family: Times;font-size: 20px">a) 
In order for a steady state to occur, the net flux across the boundaries must be $0$, in particular
      $$\oint_{\text{boundary}} \nabla u\cdot \hat{n} ds=0$$
      Since the flux across three of the sides is $0$, we may write
      $$\int^L_0f(x)dx=0$$
##### <h2 style="font-family: Times;font-size: 20px">b)
 We seek product solutions of the form
      $$u_p(x,y)=X(x)Y(y)$$
      Plugging this into the PDE yields
      $$X''(x)Y(y)+X(x)Y''(y)=0$$
      We may therefore obtain the following system of ODEs:

$$
\left\{
\begin{aligned}
X''(x) &= -\lambda X(x) \\
Y''(y) &= \lambda Y(y)
\end{aligned}
\right.
$$
For the first ODE, we have the boundary conditions $X'(0)=X'(L)=0$, which yields solutions of the form
      $$X(x)=\sum_{n=0}^{\infty}A_n\cos(\frac{n\pi x}{L})$$
      where $A_n$ are constants to be determined.
      For $Y''(y)=\lambda Y(y)$, we have solutions written in terms of hyperbolic trigs:
      $$c_1\cosh(\frac{n\pi y}{L})+c_2\sinh(\frac{n\pi y}{L})$$
      plugging in the boundary condition $Y'(0)=0$ yields
      $$\frac{n\pi}{L}\left(c_1\sinh(\frac{n\pi (0)}{L})+c_2\cosh(\frac{n\pi (0)}{L})\right)=0$$
      which yields $c_2=0$. We now combine these to give the general solution
  $$u(x,y)=\sum_{n=0}^\infty \tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L})$$
Now we deal with the inhomogeneity, plugging in $\frac{\partial u }{\partial y}(x,H)$ yields
$$ f(x)=\sum_{n=0}^\infty \frac{n \pi}{L}\tilde{A}_n\cos(\frac{n\pi x}{L})\sinh(\frac{n \pi H}{L})$$
For $n>0$ the solvability condition doesn't really matter! In that case we have
$$\frac{n \pi}{L}\tilde{A}_n\sinh(\frac{n \pi H}{L})=\frac{2}{L}\int^L_0\cos(\frac{n\pi x}{L})f(x)dx$$
BUT, for $n=0$, we see that
$$\frac{(0)\pi}{L}\tilde{A}_0\sinh(\frac{(0)\pi H}{L})=0=\frac{1}{L}\int_0^Lf(x)dx$$
this suggests that the boundary condition "forces" that 
$$\int_0^Lf(x)dx=0$$
which is precisely the solvability condition we proposed in part (a). Assuming that

$$\frac{1}{L}\int_0^Lf(x)dx=0$$
we have the following solution
$$u(x,y)=\tilde{A}_0+\sum_{n=1}^\infty \tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L})$$
where
$$\tilde{A}_n=\frac{2}{n\pi \sinh(\frac{n\pi H}{L})}\int^L_0\cos(\frac{n\pi x}{L})f(x)dx$$
However, this solution is unique up to an arbitrary additive constant, as there's no way to determine $\tilde{A}_0$, since any $\tilde{A}_0$ would satisfy
$$\frac{(0)\pi}{L}\tilde{A}_0\sinh(\frac{(0)\pi H}{L})=0$$
##### <h2 style="font-family: Times;font-size: 20px">c)
 If we treat this as the equilibrium state of a time dependent heat equation on a rectangle with insulated boundaries, then, by the conservation of heat energy:
$$\iint_Ru(x,y,0)dA=\iint_R \tilde{A}_0+\sum_{n=1}^\infty \tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L}) dA$$
which leads to the following
$$
\begin{aligned}
\iint_R g(x,y)dA
&=\iint_R \tilde{A}_0+\sum_{n=1}^\infty \tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L}) dA\\
&=\tilde{A}_0HL+\sum_{n=1}^\infty \iint_R\tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L}) dA\\
&=\tilde{A}_0HL
\end{aligned}
$$

Notice that this is because

$$
\begin{aligned}
\iint_R\tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L}) dA
&=\int_0^H\int_0^L\tilde{A}_n\cos(\frac{n\pi x}{L})\cosh(\frac{n \pi y}{L})dxdy\\
&=\int_0^H\left[\frac{L}{n\pi}\tilde{A}_n\sin(\frac{n\pi x}{L})\cosh(\frac{n\pi y}{L})\right]_{0}^Ldy\\
&=0
\end{aligned}
$$

So

$$
\tilde{A}_0=\frac{1}{HL}\int_0^H\int_0^Lg(x,y)dxdy
$$

### <h2 style="font-family: Times;font-size: 20px">*2.5.5*
Solve Laplace's equation outside a circular disk $(r\geq a)$ subject to the boundary conditions
### <h2 style="font-family: Times;font-size: 20px;"> a)
$$u(a,\theta)=\ln 2+4\cos3\theta$$
### <h2 style="font-family: Times;font-size: 20px;"> b)
$$u(a,\theta)=f(\theta)$$
### <h2 style="font-family: Times;font-size: 20px">*Solution.*
Recall that Laplace's equation can be written in the following form in polar coordinates

$$
\nabla^2 u
=
\frac{1}{r}\frac{\partial}{\partial r}\left(r \frac{\partial u}{\partial r}\right)
+
\frac{1}{r^2}\frac{\partial^2 u}{\partial \theta^2}
=
0
$$

We seek solutions of the form

$$
G(r)\phi(\theta)
$$

Plugging this into the PDE yields

$$
\frac{r}{G}\frac{\partial}{\partial r}\left(r \frac{\partial G}{\partial r}\right)
=
-\frac{1}{\phi}\frac{\partial^2 \phi}{\partial \theta^2}
=
\lambda
$$

Which yields the following system of ODEs:

$$
\left\{
\begin{aligned}
\phi''(\theta) = -\lambda \phi(\theta)\\
r^2 G''(r) + r G'(r) - \lambda G = 0 
\end{aligned}
\right.
$$

Since we assume perfect thermal contact inside the ring, we have the boundary conditions

$$
\phi(-\pi) = \phi(\pi)
$$

$$
\phi'(-\pi) = \phi'(\pi)
$$

Therefore, for

$$
\phi''(\theta) = -\lambda \phi(\theta)
$$

we have solutions of the form

$$
c_1 \cos(n\theta) + c_2 \sin(n\theta)
$$

with eigenvalue $\lambda = n^2$, where $n = 0,1,2,\dots$.

Now we consider the ODE

$$
r^2 G''(r) + r G'(r) - \lambda G = 0
$$

Using the fact that this is equidimensional, we seek solutions of the form $r^p$. Plugging this into the ODE yields

$$
r^2 p(p-1) r^{p-2}
+
r p r^{p-1}
-
\lambda r^p
=
0
$$

$$
r^p \left( p(p-1) + p - \lambda \right) = 0
$$

Therefore

$$
p(p-1) + p - \lambda = 0
$$

Solving for $p$ yields

$$
p^2 = \lambda = n^2
$$

So

$$
p = \pm n
$$

For $n \geq 1$, the general solution is

$$
c_1 r^n + c_2 r^{-n}
$$

For $n = 0$, we have

$$
c_1 + c_2 \ln(r)
$$

Since we want $u(r,\theta)$ to remain finite as $r \to \infty$, we throw out $r^n$ and $\ln(r)$, so the general solution is

$$
\sum_{n=0}^{\infty} A_n r^{-n} \cos(n\theta)
+
\sum_{n=1}^{\infty} B_n r^{-n} \sin(n\theta)
$$

Now we solve for the particular solutions using the given BCs:
### <h2 style="font-family: Times;font-size: 20px;"> a)
Applying the boundary condition $u(a,\theta)=\ln 2 + 4\cos(3\theta)$ yields

$$
\ln 2 + 4\cos(3\theta)
=
\sum_{n=0}^{\infty} A_n r^{-n}\cos(n\theta)
+
\sum_{n=1}^{\infty} B_n r^{-n}\sin(n\theta)
$$

By inspection we see that

$$
A_0 = \ln 2
$$

and

$$
A_3 a^{-3} \cos(3\theta) = 4\cos(3\theta)
$$

Equating the coefficients, we have

$$
A_3 = 4a^3
$$

and all other constants are $0$. So the solution is

$$
u(r,\theta)
=
\ln 2
+
4a^3 r^{-3} \cos(3\theta)
$$
### <h2 style="font-family: Times;font-size: 20px;"> b)
Applying the boundary condition $u(a,\theta)=f(\theta)$ yields

$$
f(\theta)
=
\sum_{n=0}^{\infty} A_n a^{-n}\cos(n\theta)
+
\sum_{n=1}^{\infty} B_n a^{-n}\sin(n\theta)
$$

Using orthogonality relations, we have

$$
A_0
=
\frac{1}{2\pi}
\int_{-\pi}^{\pi} f(\theta)\, d\theta
$$

$$
A_n a^{-n}
=
\frac{1}{\pi}
\int_{-\pi}^{\pi} f(\theta)\cos(n\theta)\, d\theta
$$

$$
B_n a^{-n}
=
\frac{1}{\pi}
\int_{-\pi}^{\pi} f(\theta)\sin(n\theta)\, d\theta
$$

So

$$
A_n
=
a^n
\frac{1}{\pi}
\int_{-\pi}^{\pi} f(\theta)\cos(n\theta)\, d\theta
$$

$$
B_n
=
a^n
\frac{1}{\pi}
\int_{-\pi}^{\pi} f(\theta)\sin(n\theta)\, d\theta
$$

### <h2 style="font-family: Times;font-size: 20px">*2.5.6*
For Laplace's equation inside a circular disk $(r \leq a)$, show that

$$
u(r,\theta)
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta})
\left(
-\frac{1}{2}
+
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
\cos\big(n(\theta-\bar{\theta})\big)
\right)
d\bar{\theta}
$$
### <h2 style="font-family: Times;font-size: 20px">*Solution.*
We know that

$$
u(r,\theta)
=
\sum_{n=0}^{\infty} A_n r^n \cos(n\theta)
+
\sum_{n=0}^{\infty} B_n r^n \sin(n\theta)
$$

where

$$
A_0
=
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta}) \, d\bar{\theta}
$$

$$
A_n
=
a^{-n}
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta}) \cos(n\bar{\theta}) \, d\bar{\theta}
$$

$$
B_n
=
a^{-n}
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta}) \sin(n\bar{\theta}) \, d\bar{\theta}
$$

Plugging these into the solution yields

$$
\begin{aligned}
u(r,\theta)
&=
\sum_{n=0}^{\infty}
\left(
a^{-n}
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta}) \cos(n\bar{\theta}) \, d\bar{\theta}
\right)
r^n \cos(n\theta) \\
&\quad +
\sum_{n=0}^{\infty}
\left(
a^{-n}
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta}) \sin(n\bar{\theta}) \, d\bar{\theta}
\right)
r^n \sin(n\theta) \\
&\quad -
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta}) \, d\bar{\theta}
\\[6pt]
&=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta})
\left(
-\frac{1}{2}
+
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
\big(
\cos(n\bar{\theta})\cos(n\theta)
+
\sin(n\bar{\theta})\sin(n\theta)
\big)
\right)
d\bar{\theta}
\\[6pt]
&=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta})
\left(
-\frac{1}{2}
+
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
\cos\big(n(\bar{\theta}-\theta)\big)
\right)
d\bar{\theta}
\end{aligned}
$$

It is easy to see that

$$
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
\cos\big(n(\bar{\theta}-\theta)\big)
$$

is simply the real part of the geometric series

$$
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
e^{i n(\bar{\theta}-\theta)}
$$

If we assume that $r<a$, then

$$
\left|
\left(\frac{r}{a}\right)^n
e^{i n(\bar{\theta}-\theta)}
\right|
<
1
$$

and by the Weierstrass M-test the geometric series converges absolutely and uniformly inside the disk, so we may use the formula for the sum of an infinite geometric series and see that

$$
\begin{aligned}
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
e^{i n(\bar{\theta}-\theta)}
&=
\frac{1}{1-\frac{r}{a}e^{i(\bar{\theta}-\theta)}}
\\[6pt]
&=
\frac{1}{1-\frac{r}{a}\cos(\bar{\theta}-\theta)
-i\frac{r}{a}\sin(\bar{\theta}-\theta)}
\\[6pt]
&=
\frac{1-\frac{r}{a}\cos(\bar{\theta}-\theta)
+i\frac{r}{a}\sin(\bar{\theta}-\theta)}
{1-2\frac{r}{a}\cos(\bar{\theta}-\theta)
+\left(\frac{r}{a}\right)^2}
\end{aligned}
$$

Taking the real part gives

$$
\begin{aligned}
\sum_{n=0}^{\infty}
\left(\frac{r}{a}\right)^n
\cos\big(n(\bar{\theta}-\theta)\big)
&=
\frac{1-\frac{r}{a}\cos(\bar{\theta}-\theta)}
{1-2\frac{r}{a}\cos(\bar{\theta}-\theta)
+\left(\frac{r}{a}\right)^2}
\\[6pt]
&=
\frac{a^2-ar\cos(\bar{\theta}-\theta)}
{a^2-2ar\cos(\bar{\theta}-\theta)+r^2}
\end{aligned}
$$

Hence

$$
\begin{aligned}
u(r,\theta)
&=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta})
\left(
-\frac{1}{2}
+
\frac{a^2-ar\cos(\bar{\theta}-\theta)}
{a^2-2ar\cos(\bar{\theta}-\theta)+r^2}
\right)
d\bar{\theta}
\\[6pt]
&=
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
f(\bar{\theta})
\left(
\frac{a^2-r^2}
{a^2-2ar\cos(\bar{\theta}-\theta)+r^2}
\right)
d\bar{\theta}
\\[6pt]
&=
\frac{a^2-r^2}{2\pi}
\int_{-\pi}^{\pi}
\frac{f(\bar{\theta})}
{a^2-2ar\cos(\bar{\theta}-\theta)+r^2}
\, d\bar{\theta}
\end{aligned}
$$

as required.

### <h2 style="font-family: Times;font-size: 20px">*2.5.8*
Solve Laplace's equation inside a semicircle of radius $a$ subject to the boundary conditions
### <h2 style="font-family: Times;font-size: 20px;"> a)
$u=0$ on the diameter and $u(a,\theta)=g(\theta)$
### <h2 style="font-family: Times;font-size: 20px;"> b)
$u(a,\theta)=g(\theta)$
### <h2 style="font-family: Times;font-size: 20px">*Solution.*
### <h2 style="font-family: Times;font-size: 20px;"> a)
We seek solutions of the form

$$
G(r)\phi(\theta)
$$

Using separation of variables we obtain the following system of ODEs:

$$
\left\{
\begin{aligned}
\phi''(\theta) &= -\lambda \phi(\theta) \\
r^2 G''(r) + r G'(r) - \lambda G(r) &= 0
\end{aligned}
\right.
$$

For the first ODE, we have the boundary conditions

$$
\phi(0) = \phi(\pi) = 0
$$

Using the reference table, we know that the solutions are of the form

$$
\sin\left(\frac{n\pi \theta}{\pi}\right)
=
\sin(n\theta)
$$

where

$$
n = 1,2,\dots
$$

For

$$
r^2 G''(r) + r G'(r) - \lambda G(r) = 0
$$

we seek solutions of the form $r^p$. Plugging this into the ODE yields

$$
r^2 p(p-1) r^{p-2}
+
r p r^{p-1}
-
\lambda r^p
=
0
$$

$$
p^2 - p + p - \lambda = 0
$$

$$
p^2 = \lambda = n^2
$$

So

$$
p = \pm n
$$

where $n = 1,2,\dots$. We therefore have solutions of the form

$$
c_1 r^n + c_2 r^{-n}
$$

Since we want our solution to remain bounded as $r \to 0$, we throw out $r^{-n}$, so the general solution is

$$
\sum_{n=1}^{\infty} B_n r^n \sin(n\theta)
$$

Applying the boundary condition yields

$$
g(\theta)
=
\sum_{n=1}^{\infty} B_n a^n \sin(n\theta)
$$

Using the reference table (or standard orthogonality relations) yields

$$
a^n B_n
=
\frac{2}{\pi}
\int_0^{\pi}
g(\theta)\sin(n\theta)\, d\theta
$$

So

$$
B_n
=
\frac{2}{\pi}
\int_0^{\pi}
a^{-n} g(\theta)\sin(n\theta)\, d\theta
$$
### <h2 style="font-family: Times;font-size: 20px;"> b)
For an insulated diameter, we have the same system of ODEs but different boundary conditions. That is, we want to solve

$$
\left\{
\begin{aligned}
\phi''(\theta) = -\lambda \phi(\theta) \\
r^2 G''(r) + r G'(r) - \lambda G(r) &= 0
\end{aligned}
\right.
$$

subject to the boundary condition

$$
\phi'(0) = \phi'(\pi) = 0
$$

Using the reference table, we have

$$
\phi_n(\theta) = \cos(n\theta)
$$

where $n = 0,1,2,\dots$.

Recycling our previous work, for $G(r)$ we have solutions of the form

$$
c_1 r^n + c_2 r^{-n}
$$

where $n \geq 1$, and

$$
c_1 + c_2 \ln(r)
$$

where $n = 0$.

Similarly, we throw out $r^{-n}$ and $\ln(r)$ to make sure that the solution remains finite as $r \to 0$. So the general solution is

$$
u(r,\theta)
=
\sum_{n=0}^{\infty}
A_n r^n \cos(n\theta)
$$

Now we solve for the constants. Plugging in $u(a,\theta)=g(\theta)$ yields

$$
g(\theta)
=
\sum_{n=0}^{\infty}
A_n a^n \cos(n\theta)
$$

Using the reference table (or standard orthogonality relations) yields

$$
A_0
=
\frac{1}{\pi}
\int_{0}^{\pi}
g(\theta)\, d\theta
$$

and for $n \geq 1$, we have

$$
A_n a^n
=
\frac{2}{\pi}
\int_{0}^{\pi}
g(\theta)\cos(n\theta)\, d\theta
$$

which yields

$$
A_n
=
a^{-n}
\frac{2}{\pi}
\int_{0}^{\pi}
g(\theta)\cos(n\theta)\, d\theta
$$


### <h2 style="font-family: Times;font-size: 20px">*2.5.10*
Solve Laplace's equation on a circular annulus subject to the boundary conditions

$$
u(a,\theta) = f(\theta)
$$

$$
u(b,\theta) = g(\theta)
$$
### <h2 style="font-family: Times;font-size: 20px">*Solution.*
Using the method of separation of variables, we obtain the following system of ODEs

$$
\left\{
\begin{aligned}
\phi''(\theta) &= -\lambda \phi(\theta) \\
r^2 G''(r) + r G'(r) - \lambda G(r) &= 0
\end{aligned}
\right.
$$

Again, assuming perfect thermal contact, we have the conditions

$$
\phi(-\pi) = \phi(\pi)
$$

$$
\phi'(-\pi) = \phi'(\pi)
$$

Therefore, for

$$
\phi''(\theta) = -\lambda \phi(\theta)
$$

we have solutions of the form

$$
c_1 \cos(n\theta) + c_2 \sin(n\theta)
$$

with eigenvalue $\lambda = n^2$, where $n = 0,1,2,\dots$.

Now we consider the ODE

$$
r^2 G''(r) + r G'(r) - \lambda G = 0
$$

For $n \geq 1$, the general solution is

$$
c_1 r^n + c_2 r^{-n}
$$

For $n = 0$, we have

$$
c_1 + c_2 \ln(r)
$$

Since we are solving Laplace’s equation on a bounded annulus, we do not discard any solutions. The general solution is

$$
\begin{aligned}
u(r,\theta)
&=
\sum_{n=1}^{\infty} r^n B_n \sin(n\theta)
+
\sum_{n=1}^{\infty} r^n A_n \cos(n\theta)
+
\sum_{n=1}^{\infty} r^{-n} \tilde{B}_n \sin(n\theta)
+
\sum_{n=1}^{\infty} r^{-n} \tilde{A}_n \cos(n\theta)
+
A_0 \ln r
+
\tilde{A}_0
\\[6pt]
&=
\sum_{n=1}^{\infty}
(r^n B_n + r^{-n} \tilde{B}_n)\sin(n\theta)
+
\sum_{n=1}^{\infty}
(r^n A_n + r^{-n} \tilde{A}_n)\cos(n\theta)
+
(A_0 \ln r + \tilde{A}_0)
\end{aligned}
$$

Applying the boundary condition at $r=a$ yields

$$
\begin{aligned}
f(\theta)
&=
\sum_{n=1}^{\infty}
(a^n B_n + a^{-n} \tilde{B}_n)\sin(n\theta)
+
\sum_{n=1}^{\infty}
(a^n A_n + a^{-n} \tilde{A}_n)\cos(n\theta)
+
(A_0 \ln a + \tilde{A}_0)
\end{aligned}
$$

Using orthogonality relations gives

$$
a^n B_n + a^{-n} \tilde{B}_n
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\theta)\sin(n\theta)\, d\theta
$$

$$
a^n A_n + a^{-n} \tilde{A}_n
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
f(\theta)\cos(n\theta)\, d\theta
$$

$$
A_0 \ln a + \tilde{A}_0
=
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
f(\theta)\, d\theta
$$

Similarly, applying the boundary condition at $r=b$ gives

$$
b^n B_n + b^{-n} \tilde{B}_n
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
g(\theta)\sin(n\theta)\, d\theta
$$

$$
b^n A_n + b^{-n} \tilde{A}_n
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
g(\theta)\cos(n\theta)\, d\theta
$$

$$
A_0 \ln b + \tilde{A}_0
=
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
g(\theta)\, d\theta
$$

Solving for $B_n$, we obtain

$$
(a^{2n} - b^{2n}) B_n
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
(a^n f(\theta) - b^n g(\theta))
\sin(n\theta)\, d\theta
$$

$$
B_n
=
\frac{1}{\pi (a^{2n} - b^{2n})}
\int_{-\pi}^{\pi}
(a^n f(\theta) - b^n g(\theta))
\sin(n\theta)\, d\theta
$$

Similarly,

$$
(a^{-2n} - b^{-2n}) \tilde{B}_n
=
\frac{1}{\pi}
\int_{-\pi}^{\pi}
(a^{-n} f(\theta) - b^{-n} g(\theta))
\sin(n\theta)\, d\theta
$$

$$
\tilde{B}_n
=
\frac{1}{\pi (a^{-2n} - b^{-2n})}
\int_{-\pi}^{\pi}
(a^{-n} f(\theta) - b^{-n} g(\theta))
\sin(n\theta)\, d\theta
$$

For $n=0$, we have

$$
A_0 (\ln a - \ln b)
=
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
\big(f(\theta) - g(\theta)\big)\, d\theta
$$

$$
A_0
=
\frac{1}{2\pi \ln\!\left(\frac{a}{b}\right)}
\int_{-\pi}^{\pi}
\big(f(\theta) - g(\theta)\big)\, d\theta
$$

Finally,

$$
\begin{aligned}
\tilde{A}_0
&=
\frac{1}{2\pi}
\int_{-\pi}^{\pi}
\frac{(\ln a) g(\theta) - (\ln b) f(\theta)}
{\ln a - \ln b}
\, d\theta
\end{aligned}
$$

### <h2 style="font-family: Times;font-size: 20px">*2.5.12*
Using the maximum principles for Laplace's equation, prove that the solutions of Poisson's equation, $\nabla^2u=g(\mathbf{x})$, subject to $u=f(\mathbf{x})$ on the boundary, is unique.
### <h2 style="font-family: Times;font-size: 20px">*Proof.*
To be extra rigorous, we first establish the fact that

$$
\nabla^2 = \sum_{i=1}^n \frac{\partial^2}{\partial x_i^2}
$$

is a linear operator on $C^2(\mathbb{R}^n)$, the space of twice differentiable functions from $\mathbb{R}^n$ to $\mathbb{R}$.

Let $f,g \in C^2(\mathbb{R}^n)$. Then

$$
\begin{aligned}
\nabla^2(f+g)
&= \sum_{i=1}^n \frac{\partial^2}{\partial x_i^2}(f+g) \\
&= \sum_{i=1}^n \frac{\partial}{\partial x_i}
\frac{\partial}{\partial x_i}(f+g) \\
&= \sum_{i=1}^n \frac{\partial}{\partial x_i}
\left( \frac{\partial f}{\partial x_i}
+ \frac{\partial g}{\partial x_i} \right)
\quad \text{(linearity of differentiation)} \\
&= \sum_{i=1}^n
\left(
\frac{\partial^2 f}{\partial x_i^2}
+
\frac{\partial^2 g}{\partial x_i^2}
\right) \\
&= \sum_{i=1}^n \frac{\partial^2 f}{\partial x_i^2}
+
\sum_{i=1}^n \frac{\partial^2 g}{\partial x_i^2} \\
&= \nabla^2 f + \nabla^2 g.
\end{aligned}
$$

Similarly, let $\lambda \in \mathbb{R}$ be arbitrary. Then

$$
\begin{aligned}
\nabla^2(\lambda f)
&= \sum_{i=1}^n \frac{\partial^2}{\partial x_i^2}(\lambda f) \\
&= \sum_{i=1}^n \frac{\partial}{\partial x_i}
\frac{\partial}{\partial x_i}(\lambda f) \\
&= \sum_{i=1}^n \frac{\partial}{\partial x_i}
\left( \lambda \frac{\partial f}{\partial x_i} \right) \\
&= \sum_{i=1}^n
\lambda \frac{\partial^2 f}{\partial x_i^2} \\
&= \lambda
\sum_{i=1}^n \frac{\partial^2 f}{\partial x_i^2} \\
&= \lambda \nabla^2 f.
\end{aligned}
$$

Thus, $\nabla^2$ is a linear operator on $C^2(\mathbb{R}^n)$.

Now suppose $u_1$ and $u_2$ are both solutions to Poisson's equation with the same boundary condition. Then

$$
\nabla^2 u_1 = g(\mathbf{x}),
\qquad
\nabla^2 u_2 = g(\mathbf{x}),
$$

and on the boundary,

$$
u_1 = f(\mathbf{x}),
\qquad
u_2 = f(\mathbf{x}).
$$

Subtracting the equations gives

$$
\nabla^2 u_1 - \nabla^2 u_2 = 0.
$$

By linearity of $\nabla^2$,

$$
\nabla^2 (u_1 - u_2) = 0.
$$

Moreover,

$$
u_1 - u_2 = 0
$$

on the boundary. Therefore $u_1 - u_2$ solves Laplace’s equation with zero boundary data.

By the maximum principle, and since $u_1 - u_2$ is zero on the boundary, we have

$$
0 \le |u_1(\mathbf{x}) - u_2(\mathbf{x})| \le 0.
$$

Hence

$$
u_1(\mathbf{x}) - u_2(\mathbf{x}) = 0
$$

throughout the domain, so

$$
u_1 = u_2.
$$

Therefore, the solution to Poisson’s equation is uniquely determined by the boundary condition.
